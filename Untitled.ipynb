# Diagrama de persistencia, Homologia persistente

Paso 1: Generar la Serie de Tiempo

Generamos una serie de tiempo y(t)y(t) que es una suma de una funci√≥n sinusoidal y ruido aleatorio.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar una serie de tiempo con una se√±al sinusoidal y ruido
t = np.linspace(0, 10, 500)
y = np.cumsum(np.random.normal(size=t.shape))

plt.plot(t, y)
plt.title("Serie de Tiempo")
plt.xlabel("Tiempo")
plt.ylabel("Amplitud")
plt.show()
```

Paso 2: Construcci√≥n del Espacio de Fase

Para aplicar la homolog√≠a persistente, primero construimos el espacio de fase mediante la t√©cnica de retardo de tiempo.

```python
from sklearn.neighbors import KDTree

# Funci√≥n para reconstruir el espacio de fase
def delay_embedding(data, delay, dimension):
    n_points = len(data) - (dimension - 1) * delay
    embedded_data = np.zeros((n_points, dimension))
    for i in range(dimension):
        embedded_data[:, i] = data[i * delay : i * delay + n_points]
    return embedded_data

# Par√°metros de retardo de tiempo y dimensi√≥n de incrustaci√≥n
delay = 5
dimension = 3

# Construcci√≥n del espacio de fase
embedded_data = delay_embedding(y, delay, dimension)

plt.scatter(embedded_data[:, 0], embedded_data[:, 1], s=1)
plt.title("Espacio de Fase Reconstruido")
plt.xlabel("Eje 1")
plt.ylabel("Eje 2")
plt.show()
```

Paso 3: Calcular la Homolog√≠a Persistente

Aplicamos homolog√≠a persistente a los datos incrustados en el espacio de fase.

```python
from ripser import ripser
from persim import plot_diagrams

# Calcular la homolog√≠a persistente
diagrams = ripser(embedded_data)['dgms']

# Graficar el diagrama de persistencia
plot_diagrams(diagrams, show=True)
plt.title("Diagrama de Persistencia")
plt.show()
```

Interpretaci√≥n del Diagrama de Persistencia

    Puntos cerca de la diagonal: Representan caracter√≠sticas de corta duraci√≥n (ruido).
    Puntos lejos de la diagonal: Representan caracter√≠sticas persistentes, como ciclos que corresponden a la estructura peri√≥dica de la se√±al sinusoidal.

Aplicaciones

    Identificaci√≥n de patrones: La homolog√≠a persistente puede revelar la periodicidad y otras estructuras recurrentes en la serie de tiempo.
    An√°lisis de ruido: Permite distinguir entre ruido y caracter√≠sticas significativas en la serie de tiempo.

```python
from persim import PersImage
# Graficar el Persistence Landscape
persim_image = PersImage(spread=1, pixels=[200, 200])
img = persim_image.transform(diagrams[1])  # Usamos H1 (ciclos)

plt.imshow(img, cmap='hot', interpolation='nearest')
plt.title("Persistence Landscape")
plt.colorbar()
plt.show()
```

```python
def calcular_dimension_fractal(img):
    def box_count(img, k):
        S = np.add.reduceat(
            np.add.reduceat(img, np.arange(0, img.shape[0], k), axis=0),
            np.arange(0, img.shape[1], k), axis=1)
        return len(np.where((S > 0) & (S < k*k))[0])

    # Convertir la imagen a un binario (blanco y negro)
    img = (img > img.mean()).astype(np.uint8)
    
    # Tamanos de caja
    sizes = np.array([2, 4, 8, 16, 32, 64])
    counts = []

    for size in sizes:
        counts.append(box_count(img, size))

    # Ajuste lineal para encontrar la dimensi√≥n fractal
    coeffs = np.polyfit(np.log(sizes), np.log(counts), 1)
    return -coeffs[0]
```

```python
plt.figure(figsize=(18,5))
for i,j,k in zip([altocumulos,cirrocumulos,cumulos,cumulonimbos,estratocumulo],
                 ["altocumulos","cirrocumulos","cumulos","cumulonimbos","estratocumulo"],
                [1,2,3,4,5]):
    plt.subplot(1,5,k)
    plt.imshow(i)
    plt.title(j+"={}".format(np.round(calcular_dimension_fractal(i),2)))
print("dimension fractal")
plt.show()
```
